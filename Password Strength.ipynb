As we can see,Strength = 1 has a lot of values compared to 0 and 2. Hence if we create a ML model, it will be biased towards strength = 1. Its a type of imbalanced data set. And as we can see, in this data set, Password is an independent feature and strength is the dependent feature. So we'll have to seperate both of them. For that, I am going to make an array

```python
password_arr = np.array(data)
```

```python
password_arr
```

For Randomness, shuffling the array for robustness

```python
import random
random.shuffle(password_arr)
```

```python
x=[labels[0] for labels in password_arr]
y=[labels[1] for labels in password_arr]
```

```python
x
```

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char("v1118714")
```

Convert strings into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

#### apply TF-IDF vectorizer on data

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```
